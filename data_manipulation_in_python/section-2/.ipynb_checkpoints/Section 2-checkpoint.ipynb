```python
df = pd.read_csv("heart.csv")
df
```

```python
data = df.to_numpy()
data
```

```python
df.head()
```

```python
print(data.dtype, data) #converts to floats
```

```python
data[0,0] = 100
df.head()  #original df hasn't changed
```

```python
df2 = df[["age", "sex", "cp"]]
data2 = df2.to_numpy()
data2[0,0] = 100
df2    

#numpy doesn't make a copy because selected columns are all ints
#need to be careful because the to_numpy() function sometimes will copy and 
#sometimes not.

#to force a copy, always call the copy method:
data3 = df2.to_numpy().copy()
data3[0,0] = 500
df2 # unchanged
```

```python
df.age.to_numpy()[0] = 700
df.head() # original data is changed
```

```python
#calcuate the mean of the age column
print (df['age'].mean(), df['age'].to_numpy().mean())   
```

## Creating DataFrames

```python
data = np.random.random(size=(5,3))
print(data)
df = pd.DataFrame(data=data, columns=["A", "B", "C"])
df
```

```python
# passing in columns as dictionaries
df = pd.DataFrame(data={"A": [1,2,3],
                        "B": ["Sam", "Alex", "John-O"]})
df
```

```python
#define datatype and passing in data as an array of tuples
dtype = [("A", np.int),("B", (np.str, 20))]
data = np.array([(1, "Sam"), (2, "Alex"), (3, "John")], dtype=dtype)
print(data)
print(data.dtype)
df = pd.DataFrame(data)
df
```

```python
#passing in rows as dictionaries
data = [{"A":1, "B": "Sam"}, {"A":2, "B": "Alex"}, {"A":3, "B": "John-o"}]
df = pd.DataFrame(data)
df
```
