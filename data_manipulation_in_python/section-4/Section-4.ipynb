### Apply, Map and Vectorized Functions

#### Apply - execute an arbitrary function on a dataframe or subsection

```python
data = np.round(np.random.normal(size=(4,3)), 2)
df = pd.DataFrame(data, columns=["A","B","C"])
data
```

```python
#apply function to every item in the df
df.apply(lambda x: 1 + x)
```

```python
# apply only to column A
df.A.apply(np.abs)
```

```python
#apply assumes a vectorized function (argument needs to be a vector)
#also make sure that the function doesn't mutate the original data
def double_if_positive(x):
    #x is a pandas series unless raw=True
    x = x.copy() #make a copy here 
    x[x>0] *= 2
    print(x)
    return x

#by default go column by column but axis=1 will go row by row
df.apply(double_if_positive, raw=True) #raw=True will use numpy arrays == better performance
```

#### User Defined Functions

```python
data2 = np.random.normal(10,2, size=(10000, 2))
df2 = pd.DataFrame(data2, columns=["x","y"])
df2
```

```python

# slow
def hypot1(x,y):
    return np.sqrt(x**2 + y**2)

h1 =[]
for index, (x,y) in df2.iterrows():
    h1.append(hypot1(x,y))

print(h1[0])
```

```python
# faster
def hypot2(row):
    return np.sqrt(row.x**2 + row.y**2)
h2 = df2.apply(hypot2, axis=1)
print(h2[0])
```

```python
#fully vectorized
def hypot3(xs, ys):
    return np.sqrt(xs**2 + ys**2)

h3 = hypot3(df2.x, df2.y)
print(h3[0])
```
